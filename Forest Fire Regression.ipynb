```python
xtrain, xtest, ytrain, ytest = train_test_split(x, y, test_size=0.3)
```

```python
def hyp(a, x, b):
    return (a*x+b)
```

```python
def error(a, x, b, y):
    e = 0
    m = len(x)
    
    for i in range(m):
        e += np.power((hyp(a,x[i],b) - y[i]), 2)
    return (1/(2*m))*e
```

```python
def step_grad(a, x, b, y, learning_rate):
    m = len(x)
    grad_a = 0
    grad_b = 0
    
    for i in range(m):
        grad_a += (1/m)*(hyp(a, x[i], b) - y[i])*x[i]
        grad_b += (1/m)*(hyp(a, x[i], b) - y[i])
        
    a = a - grad_a * learning_rate
    b = b - grad_b * learning_rate
    
    return a, b
```

```python
def descnd(init_a, x, init_b, y, learning_rate, iteration):
    e = 0
    a = init_a
    b = init_b
    
    for i in range(iteration):
        e = error(a, x, b, y)
        
        if i%1000 == 0:
            print(f"Error: {np.sqrt(e)}, a: {a}, b: {b}")
        
        a, b = step_grad(a, x, b, y, learning_rate)
    
    return a, b    
```
